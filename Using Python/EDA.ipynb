```python
cols = ['Coal_RB_4800_FOB_London_Close_USD',
        'Coal_RB_5500_FOB_London_Close_USD',
       'Coal_RB_5700_FOB_London_Close_USD',
       'Coal_RB_6000_FOB_CurrentWeek_Avg_USD',
       'Coal_India_5500_CFR_London_Close_USD']

df[cols] = df[cols].fillna(method='ffill')

```

```python
price_cols = [
    'Price_WTI',
    'Price_Brent_Oil',
    'Price_Dubai_Brent_Oil',
    'Price_Natural_Gas',
    'Price_Dutch_TTF',
    'Price_ICE',
    'Price_Mining',
    'Price_All_Share',
    'Price_ExxonMobil',
    'Price_Shenhua',
    'Price_ZAR_USD',
    'Price_Indian_en_exg_rate',
    'Price_LNG_Japan_Korea_Marker_PLATTS'
]

df[price_cols] = df[price_cols].interpolate(method='linear')
```

```python
df.loc[0, 'Price_Indian_en_exg_rate'] = df['Price_Indian_en_exg_rate'].bfill().iloc[0]
```

```python
df.isna().sum()
```

```python
df.reset_index
```

```python
df['Date'] = pd.to_datetime(df['Date'])
```

```python
df.to_excel('cleaned_data.xlsx', index=False)
```
